```python
numerical_columns = ["Battery",'Prize']

for x in numerical_columns:
    
    import numpy as np

    Q1 = np.quantile(d1[x], 0.25)
    Q3 = np.quantile(d1[x], 0.75)

    IQR = Q3 - Q1
    IQR

    upper_limit = Q3 + (1.5 * IQR)
    lower_limit = Q1 - (1.5 * IQR)

    d1.loc[d1[x] > upper_limit, x] = upper_limit
    d1.loc[d1[x] < lower_limit, x] = lower_limit
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=45)
```

```python
X_train
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)*100
```

```python
lr.predict(X_train)
```

```python

#K- Nearest neighbours
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn=KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)*100
```

```python
#Descion Tree
from sklearn import tree
```

```python
model=tree.DecisionTreeRegressor(criterion='squared_error')
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)*100
```

```python
#Random Forset
from sklearn.ensemble import RandomForestRegressor
```

```python
Rfr=RandomForestRegressor(n_estimators=110)
```

```python
Rfr.fit(X_train,y_train)
```

```python
Rfr.score(X_test,y_test)*100
```

```python
#Support Vector
from sklearn.svm import SVR
```

```python
svr=SVR()
```

```python
svr.fit(X_train,y_train)
```

```python
svr.score(X_test,y_test)*100
```
